## Data Visualization

- Matplotlib is a plotting library for the Python programming language and 
  its numerical mathematics extension NumPy.

- Data in a graphical format or in a graphical representation.
- Different types of graph representation.
    - Line graph(Linear graph)
    - Bar graph
    - Histogram
    - non-linear graph

### Line graph or Linear graph

- line graph is a type of graph which displays information as a series of data points called ‘markers’ connected by straight line segments.
- Line graphs are usually used to find relationship between two data sets on different axis; for instance X, Y.

```python
from matplotlib import pyplot as plt
plt.title("Linear Graph")
plt.plot([1,2,3],[1,2,3])
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
a = np.linspace(0, 10, 100)
b = np.exp(-a)
plt.plot(a, b)
plt.show()
```

```python
from matplotlib import pyplot as plt
x = [2, 4, 6]
y = [1, 3, 5]
plt.plot(x, y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

## Non-linear Graph

```python
from matplotlib import pyplot as plt
x = [1,2,3]
y = [7,4,5]
plt.title("Non linear graph")
plt.plot(x,y)
plt.show()
```

## Histogram
- A histogram shows the frequency on the vertical axis and the horizontal axis is another dimension.

```python
from matplotlib import pyplot as plt 
y = [1, 5, 8, 4, 2, 5, 4, 4, 8]
plt.title("Histogram")
plt.hist(y) # Function to plot histogram  
plt.show() # Function to show the plot 
```

```python
import matplotlib.pyplot as plt

population_ages = [22,55,62,45,21,22,34,42,42,4,99,102,110,120,121,122,130,111,115,112,80,75,65,54,44,43,42,48]

plt.hist(population_ages, bins=20, histtype='bar', rwidth=0.8)

plt.xlabel('Ages')
plt.ylabel('Frequency of papulation')
plt.title('Population Ages')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.bar([1,3,5,7,9],[5,2,7,8,2], label="Example one")

plt.bar([2,4,6,8,10],[8,6,2,5,6], label="Example two", color='g')
plt.legend()
plt.xlabel('bar number')
plt.ylabel('bar height')

plt.title('Bar Graph')

plt.show()
```

## Scatter Plot 
- Matplot has a built-in function to create scatterplots called scatter(). A scatter plot is a type of plot that shows the data as a collection of points.

```python
from matplotlib import pyplot as plt 
x = [5, 2, 9, 4, 7] 
y = [10, 5, 8, 4, 2]  
plt.scatter(x,y) 
  # function to show the plot 
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [3,2,3,4,5,6,7,8]
y = [5,2,4,2,1,4,5,2]

plt.scatter(x,y, label='this is label', color='k', s=25, marker="o")
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter plot')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1,2,3]
y = [5,7,4]

x2 = [1,2,3]
y2 = [10,14,12]

plt.plot(x, y, label='First Line')
plt.plot(x2, y2, label='Second Line')

plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Interesting Graph\nCheck it out')
plt.legend()
plt.show()
```

### Plot a graph between distance and velocity

```python
import matplotlib.pyplot as plt 
distance = list(range(2,11,2))
time = list(range(1,6))
plt.title("Distance vs Time")
plt.xlabel("Time")
plt.ylabel("Distance")
plt.plot(distance,time,"go",marker='^',markersize=10)
#plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as np
x = [1,2,3]
y = [6,8,10]
plt.plot(x, y)        # plot x and y using default line style and color
plt.plot(x, y,'bo')  # plot x and y using blue circle markers
plt.plot(y)           # plot y using x as index array 0..N-1
plt.plot(y, 'r+')     # ditto, but with red plusses
plt.show()
```

```python
import matplotlib.pyplot as np
x = [0.5,0.3,0.8]
y = [1,5,2]
plt.plot(x, y, 'go',markersize=4)
plt.plot(x, y, color='g', marker='o', linestyle='dotted',linewidth=4, markersize=12)
plt.show()
```

```python
import matplotlib.pyplot as plt
x1 = [10,30,50]
y1 = [2,6,7]
x2 = [3,5,8]
y2 = [8,4,2]
plt.plot(x1, y1, color='c',marker = 'D')
plt.plot(x2, y2, color='g',marker = 'v',markersize='12',markeredgecolor='b',markeredgewidth=0.5)
plt.show()
```
